```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.randn(100, 4), index=dates, columns=list("ABCD"))
df_mean = df.mean()  # 計算每一列的平均值
df_mean.plot(kind='bar', figsize=(10, 6), color=['red', 'green', 'blue', 'purple'])
plt.title('Bar Chart of Mean Values')
plt.ylabel('Mean Value')
plt.xlabel('Columns')
plt.show()
```

```python
df_mean = df_mean.abs() # Convert negative values to positive by taking the absolute value
df_mean.plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%', colors=['red', 'green', 'blue', 'purple'])
plt.title('Pie Chart of Mean Values')
plt.ylabel('')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 創建 X, Y 軸位置
x_pos = np.arange(len(df.columns))
y_pos = np.zeros_like(x_pos)

# 創建每個柱子的高度（使用 df 中的平均值）
z_pos = np.zeros_like(x_pos)
dx = np.ones_like(x_pos)  # 寬度
dy = np.ones_like(x_pos)  # 深度
dz = df_mean.values  # 高度對應每個列的平均值

# 繪製 3D 柱狀圖
ax.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, color='cyan', alpha=0.7)

# 設置標題和軸標籤
ax.set_title('3D Bar Chart of Mean Values')
ax.set_xlabel('Columns')
ax.set_ylabel('Y')
ax.set_zlabel('Mean Value')

plt.show()
```

```python
import numpy as np
from scipy import stats

# 三個班學生的成績
group1 = [85, 86, 88, 75, 78, 94]
group2 = [92, 94, 89, 96, 93, 95, 76, 50]
group3 = [78, 81, 79, 84, 77, 80]
group4 = [81,24,14,68,14]

# 使用 scipy.stats.f_oneway 進行單因子 ANOVA 檢定
f_statistic, p_value = stats.f_oneway(group1, group2, group3, group4)

# 輸出結果
print("F 統計量:", f_statistic)
print("p-value:", p_value)

# 解釋結果
alpha = 0.05  # 顯著性水平
if p_value < alpha:
    print("拒絕原假設：不同組別之間的均值存在顯著差異。") # This line was not indented correctly
else:
    print("未能拒絕原假設：不同組別之間的均值沒有顯著差異。") # This line was not indented correctly
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# 創建數據集
data = {
    'Score': [85, 86, 88, 75, 78, 94, 92, 94, 89, 96, 93, 95, 78, 81,
              79, 84, 77, 80],
    'Method': ['A', 'A', 'A', 'A', 'A', 'A', # 教學方法 A
               'B', 'B', 'B', 'B', 'B', 'B', # 教學方法 B
               'C', 'C', 'C', 'C', 'C', 'C'] # 教學方法 C
}

# 將數據轉換為 DataFrame
df = pd.DataFrame(data)

# 使用 statsmodels 中的 ols 方法進行單因子 ANOVA 分析
model = ols('Score ~ C(Method)', data=df).fit()

# 計算 ANOVA 表
anova_table = sm.stats.anova_lm(model, typ=2)

# 輸出 ANOVA 表
print(anova_table)
df
```

作業一個正式的報告

在某個虛構的城市裡，市政府決定要了解市民的健康狀況。他們通過調查收集了100位市民的健康分數（數值範圍從1到100），並希望根據這些數據找出一些規律，了解市民的健康狀況分佈情況。

市政府希望了解：

市民的健康分數是否存在明顯的集中趨勢。
是否有某種統計模式可以解釋市民健康分數之間的差異。
是否可以使用這些數據做出更有效的公共衛生政策建議。
市政府委託數據分析師進行數據視覺化，並使用統計模型來深入分析這些數據。

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy import stats

# 創建一個數據集，包含100個1到100之間的隨機數
np.random.seed(42)  # 設定隨機種子，讓結果可重現
data = np.random.randint(1, 101, size=100)

# 將數據轉換為 DataFrame
df = pd.DataFrame(data, columns=["HealthScore"])

# 1. 資料視覺化

# 繪製直方圖 (Histogram)
plt.figure(figsize=(8, 6))
sns.histplot(df["HealthScore"], kde=True, color="skyblue", bins=15)
plt.title('Distribution of Health Scores')
plt.xlabel('Health Score')
plt.ylabel('Frequency')
plt.show()

# 繪製箱型圖 (Box Plot)
plt.figure(figsize=(8, 6))
sns.boxplot(x=df["HealthScore"], color="lightgreen")
plt.title('Box Plot of Health Scores')
plt.xlabel('Health Score')
plt.show()

# 繪製散佈圖 (Scatter Plot) - 假設我們想檢查分數和一個虛構的“年齡”變數的關係
# 隨機生成一個年齡變數（1-100）
df["Age"] = np.random.randint(18, 80, size=100)

plt.figure(figsize=(8, 6))
sns.scatterplot(x="Age", y="HealthScore", data=df, color="orange")
plt.title('Scatter Plot: Health Score vs. Age')
plt.xlabel('Age')
plt.ylabel('Health Score')
plt.show()

# 2. 統計模型分析

# 計算基本的描述性統計：均值、標準差等
descriptive_stats = df.describe()
print(descriptive_stats)

# 使用皮爾森相關係數檢查 "Age" 和 "HealthScore" 之間的線性關係
corr, p_value = stats.pearsonr(df["Age"], df["HealthScore"])
print(f"Pearson correlation between Age and Health Score: {corr:.2f}")
print(f"P-value for correlation test: {p_value:.4f}")

# 構建線性回歸模型來檢查 "Age" 和 "HealthScore" 之間的關聯
X = sm.add_constant(df["Age"])  # 加入截距項
y = df["HealthScore"]
model = sm.OLS(y, X).fit()

# 輸出回歸結果
print(model.summary())
```

從數據的視覺化結果來看，市民的健康分數大致呈正態分佈，並且在中間範圍集中。
年齡和健康分數之間的相關性非常弱，並且回歸分析也表明年齡對健康分數的影響不顯著。
基於這些結果，市政府可以得出結論，年齡並不是影響市民健康狀況的主要因素，未來可以考慮其他因素（例如生活習慣、飲食等）來進一步分析市民的健康狀況。
