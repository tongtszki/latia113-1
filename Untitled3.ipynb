```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
dates = pd.date_range("20130101", periods=100)
df = pd.DataFrame(np.random.randn(100, 4), index=dates, columns=list("ABCD"))
df_mean = df.mean()  # è¨ˆç®—æ¯ä¸€åˆ—çš„å¹³å‡å€¼
df_mean.plot(kind='bar', figsize=(10, 6), color=['red', 'green', 'blue', 'purple'])
plt.title('Bar Chart of Mean Values')
plt.ylabel('Mean Value')
plt.xlabel('Columns')
plt.show()
```

```python
df_mean = df_mean.abs() # Convert negative values to positive by taking the absolute value
df_mean.plot(kind='pie', figsize=(8, 8), autopct='%1.1f%%', colors=['red', 'green', 'blue', 'purple'])
plt.title('Pie Chart of Mean Values')
plt.ylabel('')
plt.show()
```

```python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# å‰µå»º X, Y è»¸ä½ç½®
x_pos = np.arange(len(df.columns))
y_pos = np.zeros_like(x_pos)

# å‰µå»ºæ¯å€‹æŸ±å­çš„é«˜åº¦ï¼ˆä½¿ç”¨ df ä¸­çš„å¹³å‡å€¼ï¼‰
z_pos = np.zeros_like(x_pos)
dx = np.ones_like(x_pos)  # å¯¬åº¦
dy = np.ones_like(x_pos)  # æ·±åº¦
dz = df_mean.values  # é«˜åº¦å°æ‡‰æ¯å€‹åˆ—çš„å¹³å‡å€¼

# ç¹ªè£½ 3D æŸ±ç‹€åœ–
ax.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, color='cyan', alpha=0.7)

# è¨­ç½®æ¨™é¡Œå’Œè»¸æ¨™ç±¤
ax.set_title('3D Bar Chart of Mean Values')
ax.set_xlabel('Columns')
ax.set_ylabel('Y')
ax.set_zlabel('Mean Value')

plt.show()
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# å‰µå»ºæ•¸æ“šé›†
data = {
    'Score': [85, 86, 88, 75, 78, 94, 92, 94, 89, 96, 93, 95, 78, 81,
              79, 84, 77, 80],
    'Method': ['A', 'A', 'A', 'A', 'A', 'A', # æ•™å­¸æ–¹æ³• A
               'B', 'B', 'B', 'B', 'B', 'B', # æ•™å­¸æ–¹æ³• B
               'C', 'C', 'C', 'C', 'C', 'C'] # æ•™å­¸æ–¹æ³• C
}

# å°‡æ•¸æ“šè½‰æ›ç‚º DataFrame
df = pd.DataFrame(data)

# ä½¿ç”¨ statsmodels ä¸­çš„ ols æ–¹æ³•é€²è¡Œå–®å› å­ ANOVA åˆ†æž
model = ols('Score ~ C(Method)', data=df).fit()

# è¨ˆç®— ANOVA è¡¨
anova_table = sm.stats.anova_lm(model, typ=2)

# è¼¸å‡º ANOVA è¡¨
print(anova_table)
df
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from scipy import stats

# å‰µå»ºä¸€å€‹æ•¸æ“šé›†ï¼ŒåŒ…å«100å€‹1åˆ°100ä¹‹é–“çš„éš¨æ©Ÿæ•¸
np.random.seed(42)  # è¨­å®šéš¨æ©Ÿç¨®å­ï¼Œè®“çµæžœå¯é‡ç¾
data = np.random.randint(1, 101, size=100)

# å°‡æ•¸æ“šè½‰æ›ç‚º DataFrame
df = pd.DataFrame(data, columns=["HealthScore"])

# 1. è³‡æ–™è¦–è¦ºåŒ–

# ç¹ªè£½ç›´æ–¹åœ– (Histogram)
plt.figure(figsize=(8, 6))
sns.histplot(df["HealthScore"], kde=True, color="skyblue", bins=15)
plt.title('Distribution of Health Scores')
plt.xlabel('Health Score')
plt.ylabel('Frequency')
plt.show()

# ç¹ªè£½ç®±åž‹åœ– (Box Plot)
plt.figure(figsize=(8, 6))
sns.boxplot(x=df["HealthScore"], color="lightgreen")
plt.title('Box Plot of Health Scores')
plt.xlabel('Health Score')
plt.show()

# ç¹ªè£½æ•£ä½ˆåœ– (Scatter Plot) - å‡è¨­æˆ‘å€‘æƒ³æª¢æŸ¥åˆ†æ•¸å’Œä¸€å€‹è™›æ§‹çš„â€œå¹´é½¡â€è®Šæ•¸çš„é—œä¿‚
# éš¨æ©Ÿç”Ÿæˆä¸€å€‹å¹´é½¡è®Šæ•¸ï¼ˆ1-100ï¼‰
df["Age"] = np.random.randint(18, 80, size=100)

plt.figure(figsize=(8, 6))
sns.scatterplot(x="Age", y="HealthScore", data=df, color="orange")
plt.title('Scatter Plot: Health Score vs. Age')
plt.xlabel('Age')
plt.ylabel('Health Score')
plt.show()

# 2. çµ±è¨ˆæ¨¡åž‹åˆ†æž

# è¨ˆç®—åŸºæœ¬çš„æè¿°æ€§çµ±è¨ˆï¼šå‡å€¼ã€æ¨™æº–å·®ç­‰
descriptive_stats = df.describe()
print(descriptive_stats)

# ä½¿ç”¨çš®çˆ¾æ£®ç›¸é—œä¿‚æ•¸æª¢æŸ¥ "Age" å’Œ "HealthScore" ä¹‹é–“çš„ç·šæ€§é—œä¿‚
corr, p_value = stats.pearsonr(df["Age"], df["HealthScore"])
print(f"Pearson correlation between Age and Health Score: {corr:.2f}")
print(f"P-value for correlation test: {p_value:.4f}")

# æ§‹å»ºç·šæ€§å›žæ­¸æ¨¡åž‹ä¾†æª¢æŸ¥ "Age" å’Œ "HealthScore" ä¹‹é–“çš„é—œè¯
X = sm.add_constant(df["Age"])  # åŠ å…¥æˆªè·é …
y = df["HealthScore"]
model = sm.OLS(y, X).fit()

# è¼¸å‡ºå›žæ­¸çµæžœ
print(model.summary())
```
